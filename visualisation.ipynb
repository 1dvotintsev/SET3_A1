Подключим пакеты для визуализации и работы с данными

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

# Чтение данных из файла
data = pd.read_csv('results.txt', delim_whitespace=True, skiprows=1, names=['PointsCount', 'EstimatedArea', 'RelativeError'])

# Точная площадь
exact_area = 0.25 * np.pi + 1.25 * np.arcsin(0.8) - 1
```

Построим график приближенной площади от количества точек

```python

# График 1: Приближенная площадь в зависимости от количества точек
plt.figure(figsize=(10, 5))
plt.plot(data['PointsCount'], data['EstimatedArea'], label='Приближенная площадь', color='g')
plt.axhline(y=exact_area, color='r', linestyle='--', label='Точная площадь')
plt.xlabel('Количество точек')
plt.ylabel('Приближенная площадь')
plt.title('Изменение приближенной площади в зависимости от количества точек')
plt.legend()
plt.grid(True)
plt.show()
```

Построим график зависимости относительной ошибки от количества точек

```python

# График 2: Относительная ошибка в зависимости от количества точек
plt.figure(figsize=(10, 5))
plt.plot(data['PointsCount'], data['RelativeError'], label='Относительная ошибка', color='g')
plt.xlabel('Количество точек')
plt.ylabel('Относительная ошибка')
plt.title('Изменение относительной ошибки в зависимости от количества точек')
plt.legend()
plt.grid(True)
plt.show()
```

## Выводы

**Эффективность метода Монте-Карло:**  
Метод Монте-Карло дает точное приближение площади фигуры при большом числе случайных точек. Однако для достижения точных результатов необходимо использовать достаточно большое количество точек (более 10,000), иначе результаты будут иметь значительные погрешности.

**Зависимость от количества точек:**  
Мы заметили, что с увеличением числа случайных точек приближенная площадь становится все более точной, а относительное отклонение стремится к нулю. Таким образом, для достижения требуемой точности необходимо увеличивать число точек. После 10000 точек результат более чем качественный, после этой отметки не наблюдается такого экспонентциальноого улучшения точности, значения меняются в пределах одной погрешности, обусловленной стахостическим подходом

**Природа отклонений:**  
Отклонения от точного значения при малом числе точек объясняются случайностью выборки. Это характерно для метода Монте-Карло, где точность результатов зависит от числа случайных испытаний.
